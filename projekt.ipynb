```python
import pandas as pd
df = pd.read_csv('messy_data.csv')
```

```python
df
```

```python
# a. Usunięcie duplikatów
df = df.drop_duplicates()
df
```

```python
# df = df.dropna()
df
```

```python
# df.info()
df.columns = df.columns.str.strip()
numeric_columns = ['x dimension', 'y dimension', 'z dimension', 'depth', 'price', 'table']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
df['clarity'] = df['clarity'].astype("string")
df['clarity'] = df['clarity'].str.upper()
df['color'] = df['color'].astype("string")
df['color'] = df['color'].str.upper()
df['cut'] = df['cut'].astype("string")
df['cut'] = df['cut'].str.upper()
df = df.sort_values("price")
df['carat'].fillna(method='ffill', inplace=True)
df['price'].fillna(method='ffill', inplace=True)
df['table'].fillna(method='ffill', inplace=True)
df['x dimension'].fillna(method='ffill', inplace=True)
df['y dimension'].fillna(method='ffill', inplace=True)
df['z dimension'].fillna(method='ffill', inplace=True)
df['depth'].fillna(method='ffill', inplace=True)
# df = df.dropna(axis=0, subset=numeric_columns)
# df['price'] = df['price'].astype("int64")
df.info()
df
```

```python
# df = df.drop('table', axis=1)
df
```

```python
# wartosci odstajace na carat
q_low = df["carat"].quantile(0.01)
q_hi  = df["carat"].quantile(0.99)
print(q_low)
print(q_hi)
df = df[(df["carat"] < q_hi) & (df["carat"] > q_low)]
df
```

```python
#kolumna cut jako categorical
df['cut'] = df.cut.astype('category')
df['cut'].describe()
```

```python
import statsmodels.formula.api as smf
model = smf.ols(formula="price ~ carat + Q('x dimension') + Q('y dimension') + Q('z dimension') + depth + table", data=df).fit()
print(model.summary())
model = smf.ols(formula="price ~ carat + Q('x dimension') + Q('y dimension') + Q('z dimension') + depth", data=df).fit()
print(model.summary())
model = smf.ols(formula="price ~ carat + Q('x dimension') + Q('y dimension') + Q('z dimension')", data=df).fit()
print(model.summary())
model = smf.ols(formula="price ~ carat + Q('x dimension') + Q('y dimension')", data=df).fit()
print(model.summary())
model = smf.ols(formula="price ~ carat + Q('x dimension')", data=df).fit()
print(model.summary())
model = smf.ols(formula="price ~ carat", data=df).fit()
print(model.summary())
```

```python
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
model = smf.ols(formula="price ~ carat + depth + table", data=df).fit()
print(model.summary())
Y = model.predict(df[['carat', 'depth', 'table']])
print(Y)
plt.scatter(range(len(Y)), Y)
plt.plot(np.sort(Y), color='red', label='Linia regresji')
plt.show()
model = smf.ols(formula="price ~ carat + depth", data=df).fit()
Y = model.predict(df[['carat', 'depth']])
print(model.summary())
print(Y)
plt.scatter(range(len(Y)), Y)
plt.plot(np.sort(Y), color='red', label='Linia regresji')
plt.show()
model = smf.ols(formula="price ~ carat", data=df).fit()
print(model.summary())
Y = model.predict(df['carat'])
print(Y)
plt.scatter(range(len(Y)), Y)
plt.plot(np.sort(Y), color='red', label='Linia regresji')
plt.show()
```
